Simple Linear Regression
1. Calculate the Mean
2. Calulate the adjusted Mean
3. Compute the y-intercept(b0)
4. Compute the Slope(b1)
5. Predict values

```python
import numpy as np
import pandas as pd
import seaborn as sns

```

```python
x = [1,2,3,4,5]
y = [4,2,5,3,4]
location = "/content/LifeExpectancyData.csv"
data = pd.read_csv('/content/LifeExpectancyData.csv')


filter_year = 2011
f = data['Year'] == filter_year
data_filtered = data[f]
data_filtered = data_filtered.fillna(0)
```

```python
variable_x = 'GDP'
variable_y ='Life expectancy '

x = data_filtered[variable_x]

y = data_filtered[variable_y]
```

```python
def calcBeta(adjusted_x, adjusted_y, mean_x, mean_y):
  b1 = sum((adjusted_x)*(adjusted_y)) / sum((adjusted_x)**2)
  b0 = mean_y - b1*mean_x
  return b0, b1

def predictY(b0, b1, x):
  pred_y = []
  for num in x:
    pred_y.append(b0 + b1*num)
  return pred_y
```

```python
meanx = np.mean(x)
meany = np.mean(y)
adj_x = x - meanx
adj_y = y - meany
beta_0, beta_1 = (calcBeta(adj_x, adj_y, meanx, meany))
approximate_y = predictY(beta_0,beta_1, x)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,5))
sns.scatterplot(data=data_filtered, x=x, y=y, hue='Country', size='GDP', sizes=(10,200), legend=False)
sns.scatterplot(x=x, y=approximate_y)
```
